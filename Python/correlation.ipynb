```python
# checking for missing data via loop
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

With the current missing values in rating, budget and gross, there are five ways in which we could deal with this discrepancy in the dataset that is;
* Dropping the missing values
* Replacing the missing values with a specific value or a computed one
* Interpolate to estimate missing values
* Use backward/forward fill to propagate the next or previous values to fill the missing ones
* Dropping columns with all the missing values

We could also decide that the rows with missing values are not useful based on the missing feature and dropp it whole. The method we chose is dependent on the nature of our data and the importance of missing values. We can gauge the importance of fetures by observing the correlation. Since we are set to observe the correlations within our data, I saw it prudent to drop the rows with values missing on budget, rating and gross for accuracy reasons.

```python
df=df.dropna(how='any')
```

by using the all parameter, the 2% data with missing values on rating was wiped out indicating the rest of the features were missing as well. while using the all we then got rid of the rest of the rows with missing values.

```python
# confirming the rows with missing data have been eliminated
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

Mission acomplished! We no longer have missing data in our dataset. Moving on...

```python
# checking for duplicate values via loop
for col in df.columns:
    pct_missing = np.mean(df[col].duplicated())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Assigning a random numeric value to each unique categorical value
df.apply(lambda x: x.factorize()[0]).corr(method="pearson", numeric_only=True)
```

```python
corr_matrix=df.apply(lambda x: x.factorize()[0]).corr(method="pearson", numeric_only=True)
sns.heatmap(corr_matrix, annot=True)
plt.title("Complete Correlation Matrix")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()
```

```python
correlation_mat= df.apply(lambda x: x.factorize() [0]).corr(numeric_only=True)
corr_pairs = correlation_mat.unstack()
print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# looking at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
