```python
# importing libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
```

```python
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Reading the movies data
df = pd.read_csv(r'C:\Users\pc\Desktop\SQL\movies.csv')
```

```python
# previewing the data
df
```

```python
# checking for missing data via loop
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# checking for duplicate values via loop
for col in df.columns:
    pct_missing = np.mean(df[col].duplicated())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Checking the datatypes
print(df.dtypes)
```

## 1. Are there any Outliers?

```python
df.boxplot(column=['gross'])
```

```python
# Dropping duplicates
df.drop_duplicates()
```

```python
# Dropping missing values
df.dropna(axis=1, how='all')
```

```python
# Sorting the data in descending order for observation
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# regression plot of gross vs budget
sns.regplot(x="gross", y="budget", data=df)
```

```python
correlation_matrix=df.corr(method='pearson', numeric_only=True)
```

```python
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# Assigning a random numeric value to each unique categorical value
df.apply(lambda x: x.factorize()[0]).corr(method="pearson", numeric_only=True)
```

```python
corr_matrix=df.apply(lambda x: x.factorize()[0]).corr(method="pearson", numeric_only=True)
sns.heatmap(corr_matrix, annot=True)
plt.title("Complete Correlation Matrix")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()
```
